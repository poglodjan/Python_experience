# 4. function which calculate the score

```python
from sklearn.metrics import make_scorer
def _score_func(y_true, y_pred): 
    weights = np.array([[0,1,2,4],
                        [2,0,1,2],
                        [4,2,0,1],
                        [8,4,2,0]])
    confusion_matri = confusion_matrix(y_true, y_pred)
    loss = (weights*confusion_matri).sum()
    return loss

ag_LOSS_scorer = make_scorer(name='LOSS',
                             score_func=_score_func,
                             optimum=0,
                             greater_is_better=False)

#TabularPredictor(..., eval_metric=ag_LOSS_scorer)
```

```python
def _new_scorer(y_true, y_pred):
    # Defines a custom scorer function

    weights = np.array([[0,1,2,4],
                        [2,0,1,2],
                        [4,2,0,1],
                        [8,4,2,0]])
    cm = confusion_matrix(y_true, y_pred)
    loss = (weights*cm).sum()
    return loss
    
new_scorer = make_scorer(_new_scorer, greater_is_better = False)
```

# 5. Creating a pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import FunctionTransformer
from datetime import datetime

cols_to_drop = ['CashFlow2', 'CashFlow3', 'Profitability6', 'Profitability7',
       'Efficiency1', 'Efficiency2','Profitability1', 'Profitability4','Profitability8']
def func(df):
    df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y").dt.month
    return df

preprocessor = ColumnTransformer([
    ('drop', 'drop', cols_to_drop),
    ('date', FunctionTransformer(func), ["Date"]),
    ('onehot', OneHotEncoder(), ["Sector"])
], remainder='passthrough')
```

```python
X_train = pd.read_csv("train_fixed.csv", thousands=' ')
X_train = X_train.drop("Rating",axis=1)
```

```python
y_train = pd.read_csv("train_fixed.csv")
y_train = y_train.Rating
mapping = {'Low Risk': 0, 'Medium Risk': 1, 'High Risk': 2, 'Highest Risk': 3}  
y_train = y_train.replace(mapping)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.3)
```

```python
from xgboost import XGBClassifier
model_xg = Pipeline([
        ('pre', preprocessor),
        ('feature_selector', SequentialFeatureSelector(RandomForestClassifier(n_estimators=300, max_depth=25, criterion='gini', min_samples_leaf=1),n_features_to_select=20)),  # selekcja zmiennych
        ('classifier', XGBClassifier(learning_rate=0.2, max_depth=7,n_estimators=100))
    ])

model_xg.fit(X_train,y_train)
y_pred = model_xg.predict(X_test)
```
