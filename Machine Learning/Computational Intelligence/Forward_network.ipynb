```python
x = np.array(small_train['x'])
y_true = np.array(small_train['y'])

# losowe wagi
weights_input_hidden = np.random.randn(100,5)  
biases_hidden = np.random.randn(5)  
weights_hidden_output = np.random.randn(5,100)  
bias_output = np.random.randn(1,100)

# propagacja w przód
hidden_layer_input = np.dot(x, weights_input_hidden) + biases_hidden
hidden_layer_output = sigmoid(hidden_layer_input)
output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output 

y_predicted = output_layer_input

error = y_predicted - y_true
print("Sum of all errors: ", np.sum(error))
```

```python
for i in range(200):    
    learning_rate = 0.01 # Współczynnik uczenia

    # aktualizacja wag w outpucie na warstwę ukrytą
    weights_hidden_output -= learning_rate * np.outer(error, hidden_layer_output).T
    bias_output -= learning_rate * np.sum(error)
    w = np.array(np.dot(error, weights_hidden_output.T) * hidden_layer_output * (1 - hidden_layer_output))
    w=w.T*x

    # aktualizacja wag w inpucie na warstwę ukrytą
    weights_input_hidden -= w.T
    suma = np.sum(error.T * weights_hidden_output.T * hidden_layer_output * (1 - hidden_layer_output.T))
    # aktualizacja biasów dla neuronów
    biases_hidden -= learning_rate * suma

    # ponowna propagracja wprzód
    new_hidden_layer_input = np.dot(x, weights_input_hidden) + biases_hidden
    new_hidden_layer_output = sigmoid(new_hidden_layer_input)
    new_output_layer_input = np.dot(new_hidden_layer_output, weights_hidden_output) + bias_output 
    new_y_predicted = new_output_layer_input
    error = new_y_predicted - y_true
```

# 5) Creating function MLP

```python
def mlp(x_train, y_train, x_test, y_test, activation, iters, inputs_n, output_n, neurons):

    # Inicjalizacja losowych wag i biasów
    weights_input_hidden = np.random.randn(inputs_n, neurons)
    biases_hidden = np.random.randn(neurons)
    weights_hidden_output = np.random.randn(neurons, output_n)
    bias_output = np.random.randn(1, output_n)

    # Propagacja w przód dla danych treningowych
    hidden_layer_input = np.dot(x_train, weights_input_hidden) + biases_hidden
    hidden_layer_output = activation(hidden_layer_input)
    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output

    y_predicted = output_layer_input
    error = y_predicted - y_train

    for i in range(iters):
        learning_rate = 0.01

        # Aktualizacja wag w warstwie ukrytej na wyjście
        weights_hidden_output -= learning_rate * np.outer(error, hidden_layer_output).T
        bias_output -= learning_rate * np.sum(error)

        # Obliczenie wagi dla warstwy ukrytej
        w = np.dot(error, weights_hidden_output.T) * hidden_layer_output * (1 - hidden_layer_output)
        w = w.T * x_train

        # Aktualizacja wag w warstwie wejściowej na ukrytą
        weights_input_hidden -= w.T
        suma = np.sum(error.T * weights_hidden_output.T * hidden_layer_output * (1 - hidden_layer_output.T))
        biases_hidden -= learning_rate * suma

        # Ponowna propagacja w przód
        new_hidden_layer_input = np.dot(x_train, weights_input_hidden) + biases_hidden
        new_hidden_layer_output = activation(new_hidden_layer_input)
        new_output_layer_input = np.dot(new_hidden_layer_output, weights_hidden_output) + bias_output
        new_y_predicted = new_output_layer_input
        error = new_y_predicted - y_train

    print("MSE train =", np.mean(np.square(error)))

    # Propagacja w przód dla danych testowych
    hidden_layer_input_test = np.dot(x_test, weights_input_hidden) + biases_hidden.T
    hidden_layer_output_test = activation(hidden_layer_input_test)
    output_layer_input_test = np.dot(hidden_layer_output_test, weights_hidden_output) + bias_output


    difference_test = y_test - output_layer_input_test
    print("MSE test =", np.mean(np.square(difference_test)))

    return output_layer_input_test
```
